# Fuzzy Search from ES (Zauba)

```python
def elastic_name(name):
    return es.search(index='gstin',body={
  "query": {
    "match": {                                             
      "name.keyword": {                                          
        "query": name,
        "fuzziness": "AUTO",
        "operator":  "and"
         
      }
      
    }                                              
  } 
})
```
